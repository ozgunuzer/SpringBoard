Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members! */

```python
df1['guest_revenue'] = df1['guestcost'] * df1['slots']
```

```python
df1['member_revenue'] = df1['membercost'] * df1['slots']
```

```python
df1['revenue'] = 0
```

```python
df1.loc[df1['member_yes'] == 1, 'revenue'] = df1['member_revenue']
```

```python
df1.loc[df1['member_yes'] == 0, 'revenue'] = df1['guest_revenue']
```

```python
df1.head()
```

```python
result = df1.groupby(by =['name'])['revenue'].sum()
```

```python
result.sort_values()
```

```python
result[result < 1000].sort_values(ascending = False)
```
